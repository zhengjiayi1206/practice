```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, BitsAndBytesConfig

# 量化配置
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,                  # 启用 4bit 权重量化
    bnb_4bit_quant_type="nf4",          # 量化类型（推荐 "nf4"）
    bnb_4bit_use_double_quant=True,     # 双重量化进一步压缩
    bnb_4bit_compute_dtype=torch.float16  # 推理计算仍使用 float16
)

# 加载 tokenizer 和模型
tokenizer = AutoTokenizer.from_pretrained("./flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained(
    "./flan-t5-base",
    device_map="auto",                  # 自动分配 GPU / CPU
    quantization_config=bnb_config
)

# 示例推理
input_text = "translate English to German: How old are you?"
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")

outputs = model.generate(input_ids)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_id = "./flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_id)

max_mem = {
    0: "12GiB",   # 按你显存改
    1: "12GiB"
}

model = AutoModelForSeq2SeqLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",          # 让 accelerate 自动切分
    max_memory=max_mem,         # 限制每张卡的上限
)

# 推理：把输入放到第一层所在设备即可，通常是 cuda:0
text = "translate English to German: How old are you?"
inputs = tokenizer(text, return_tensors="pt").to("cuda:0")

with torch.no_grad():
    out = model.generate(**inputs, max_new_tokens=64)

print(tokenizer.decode(out[0], skip_special_tokens=True))
```
